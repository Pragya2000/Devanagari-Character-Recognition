```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
dataset = pd.read_csv("data.csv")
dataset
```

```python
dataset['character'].unique()
```

```python
dataset.duplicated()
```

```python
dataset.iloc[:,1024].value_counts()
```

```python
x = dataset.iloc[:,:-1]
y = dataset.iloc[:,-1]
```

```python
x.shape
```

```python
photo_width = 32
photo_height = 32
x_images = x.values.reshape(x.shape[0], photo_width, photo_height)
```

```python
for i in range(1,9):
    plt.subplot(240+i)
    plt.axis('off')
    plt.imshow(x_images[i-1],cmap=plt.get_cmap('gray'))
plt.show()
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.4,random_state=0,shuffle=True)
```

```python
from sklearn.tree import DecisionTreeClassifier
dtc=DecisionTreeClassifier()
dtc.fit(xtrain,ytrain)
```

```python
from sklearn import tree
tree.plot_tree(dtc)
```

```python
ypp=dtc.predict(xtest)
```

```python
from sklearn.metrics import accuracy_score
acc=accuracy_score(ypp,ytest)
```

```python
print('Accuracy for Decision Tree Classifier',acc)
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(ytest,ypp,labels=["character_02_kha","character_33_ha"])
print(cm)
#Ci,j represents the number of entries known to be in group i and predicted to be in j
```

```python
cm=confusion_matrix(ytest,ypp)
cm=cm/cm.max()
plt.figure(figsize=(10,10))
plt.matshow(cm)
plt.colorbar()
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier(n_estimators=150)
rfc.fit(xtrain,ytrain)
```

```python
ypp_rfc=rfc.predict(xtest)
print('Accuracy for Random Forest Classifier:',accuracy_score(ypp_rfc,ytest))
```

```python
cm=confusion_matrix(ytest,ypp_rfc,labels=["character_02_kha","character_33_ha"])
print(cm)
```

```python
cm=confusion_matrix(ytest,ypp_rfc)
cm=cm/cm.max()
plt.figure(figsize=(10,10))
plt.matshow(cm)
plt.colorbar()
```

```python
from sklearn.linear_model import SGDClassifier
sgd=SGDClassifier(loss="hinge", penalty="l2", max_iter=100)
sgd.fit(xtrain,ytrain)
ypp_sgd=sgd.predict(xtest)
```

```python
print('Accuracy for SGD Classifier:',accuracy_score(ypp_sgd,ytest))
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(ytest,ypp_sgd,labels=["character_02_kha","character_33_ha"])
print(cm)
```

```python
cm=confusion_matrix(ytest,ypp_sgd)
cm=cm/cm.max()
plt.figure(figsize=(10,10))
plt.matshow(cm)
plt.colorbar()
```
